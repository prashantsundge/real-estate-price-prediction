```python
import re 


def extract_parking_counts(value):
    covered =0
    open_ = 0

    if isinstance(value, str):
        matches = re.findall(r'(\d+)\s*(Covered|Open)', value)
        for count, kind in matches:
            num =int(count)
            if num > 100:
                num = 1
            if kind == "Covered":
                covered +=int(count)

            elif kind == "Open":
                open_ +=int(count)
    return pd.Series([covered, open_])

df[["Covered_parking", "Open_parking"]] = df["Car Parking"].apply(extract_parking_counts)
```

```python
import re

def parse_price(price_str):
    if isinstance(price_str, str):
        price_str = price_str.replace(",", "").strip()
        match = re.match(r"([\d.]+)\s*(Cr|Lacs?)", price_str, re.IGNORECASE)
        if match:
            value, unit = match.groups()
            value = float(value)
            if unit.lower().startswith("cr"):
                return value * 1e7
            elif unit.lower().startswith("lac"):
                return value * 1e5
    return None

df["Price_INR_Numeric"] = df["Price (INR)"].apply(parse_price)
```
